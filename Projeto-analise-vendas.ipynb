```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv', encoding='latin-1')
print(df.head())

sales_country = df.groupby('COUNTRY')['SALES'].sum()

print(sales_country)

sales_country.plot(kind='bar')
plt.title('Vendas por País')
plt.xlabel('País')
plt.ylabel('Total de Vendas')
plt.show()
```
